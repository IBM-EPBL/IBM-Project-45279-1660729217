# 6.Finding & replacing the outliers

```python
#Finding outliers

sns.boxplot(df['CreditScore'])
```

```python
#Replacing outliers

median=df.loc[df['CreditScore']>400,'CreditScore'].median() 
df["CreditScore"]=df["CreditScore"].mask(df["CreditScore"] <400,median)
sns.boxplot(df['CreditScore'])
```
